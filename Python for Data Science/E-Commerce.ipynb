##  Total Revenue generated by the E-Commerce company

```python
def compute_rev(x):
        if x > 500:
            return x*0.25
        elif x > 350:
            return x*0.15
        elif x > 100:
            return x*0.10
        else:
            return x*0.05

EC_Data['Revenue'] = EC_Data['discounted_price'].apply(compute_rev)
EC_Data.head()
```

```python
total_rev = EC_Data['Revenue'].sum()
print('The total revenue generated by the companyis', total_rev, 'Rs.')
```

# Region Wise Analysis

```python
EC_data_Add = pd.DataFrame()
EC_Data['Order_Date'] = pd.to_datetime(EC_Data.Order_Date, format='%d-%m-%Y')
EC_data_Add['Order_Month']=EC_Data['Order_Date'].dt.strftime("%Y %m")
EC_data_Add['Region']=EC_Data['Region'].copy()
EC_data_Add['Revenue']=EC_Data['Revenue'].copy()
EC_data_Add_New=pd.DataFrame(EC_data_Add.groupby(['Region','Order_Month'])['Revenue'].sum())
EC_data_Add_New
```

```python
plt.figure(figsize = (20,7))
EC_data_Add_New.sort_values(by='Order_Month', ascending=True, inplace=True)
plt.xticks(np.arange(0,14,1))
sns.lineplot(data=EC_data_Add_New,x = 'Order_Month' , y = 'Revenue', ci = False ,hue='Region');
plt.show()
```
