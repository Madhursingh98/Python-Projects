```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv('Automobile.csv')
```

```python
auto.head()
```

```python
auto.info()
```

```python
auto.describe(include='all').T
```

```python
sns.jointplot(data= auto, x='engine_size', y='horsepower', kind="hex");
```

```python
sns.lmplot(data=auto, x='Horsepower', y='Price', ci=False);
```

```python
sns.jointplot(data=auto, x='price', y='city_mpg', kind="reg" );
```

```python
plt.title('Histogram')
plt.xlim(3000,50000)
plt.ylim(0,70)
plt.xlabel('Price of cars')
plt.ylabel('Frequency')
sns.histplot(data=auto, x='price', color= 'Pink');
```

```python
sns.distplot(auto['city_mpg'])
plt.show()
```

```python
sns.distplot(auto['normalized_losses'], kde=False, rug=True)
plt.show()
```

```python
sns.jointplot(auto['engine_size'], auto['horsepower'])
plt.show()
```

```python
# kind : { "scatter" | "reg" | "resid" | "kde" | "hex" }, optional
sns.jointplot(auto['engine_size'], auto['horsepower'], kind="kde")
plt.show()
```

```python
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']])
```

```python
sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter = True)
```

```python
sns.swarmplot(auto['fuel_type'], auto['horsepower'])
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type'])
```

```python
sns.barplot(auto['body_style'], auto['horsepower'],hue=auto['engine_location'])
```

```python
sns.countplot(auto['body_style'], hue=auto['engine_location'])
```

```python
sns.pointplot(auto['fuel_system'], auto['horsepower'], hue=auto['number_of_doors'])
```

```python
sns.factorplot(x="fuel_type", 
                y="horsepower",
               hue="number_of_doors", 
               col="engine_location", 
               data=auto, 
               kind="strip")

# Various types of kind input : {point, bar, count, box, violin, strip}
```

```python
sns.lmplot(x="horsepower", y="peak_rpm", data=auto, hue="fuel_type")
```

```python
sns.scatterplot(data=auto, x='engine_size', y='horsepower', hue= 'fuel_type', style= 'fuel_type');
```

```python
sns.lmplot(data=auto, x='Horsepower', y='Price', ci=False);
```
